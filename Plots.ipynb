```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

```python
First_Method = {"Iterations": [1, 5, 10, 30, 55, 63]
                ,"Time": [3496812, 4778846, 6372739, 12661130, 26020834, 30761684]
                ,"Loss": [0.974253, 0.00276032, 1.50E-04, 8.06E-07, 6.01E-13, 3.18E-24]}	

Second_Method = {"Iterations": [1, 5, 15, 40, 70, 75]
                ,"Time": [30438956, 41705133, 161327098, 758810792, 1343082999, 1432804223]
                ,"Loss": [0.974178, 0.002892, 7.96E-05, 7.91E-05, 7.91E-05, 7.91E-05]}	

Third_Method = {"Iterations": [1, 5, 15, 40, 70, 75]
                ,"Time": [2112056, 2915795, 27441537, 49113058, 148807055, 162238582]
                ,"Loss": [0.972038, 0.0026592, 1.51E-04, 3.78E-05, 2.60E-05, 6.03E-06]}	
```

```python
plt.figure(figsize = (10, 10))

plt.plot(First_Method["Time"], First_Method["Loss"], color = "green", label = "First Method\n(number of iterations)", marker = "o", markerfacecolor = "green", markersize = 9)

for i, txt in enumerate(First_Method["Iterations"]):

    plt.annotate(txt, (First_Method["Time"][i], First_Method["Loss"][i]), fontsize=20)

plt.xlabel("Time of Execution (microseconds)", fontsize=20); plt.ylabel("Loss (value of function G)", fontsize=20)

plt.legend(prop={'size':20})

plt.savefig(".\img\Fig1.eps", format = "eps")

plt.show()
```
