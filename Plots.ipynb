```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

```python
First_Method_1 = {"Iterations": [1, 5, 10, 30, 55, 65, 79]
                ,"Time": [37061544, 50698171, 67614727, 134539082, 253640392, 331583715, 448269857]
                ,"Loss": [0.973681, 0.00275971, 1.49E-04, 4.10E-07, 1.24E-14, 3.43E-20, 4.16E-26]}	

Second_Method_1 = {"Iterations": [1, 15, 40, 70, 150, 300]
                ,"Time": [84492714, 191333912, 1567728440, 4026522498, 16017327293, 36566898622]
                ,"Loss": [0.973972, 7.24E-05, 1.46E-05, 8.59E-06, 8.59E-06, 8.21E-06]}		
```

```python
plt.figure(figsize = (10, 10))

plt.plot(First_Method_1["Time"], First_Method_1["Loss"], color = "green", label = "First Method\n(number of iterations)", marker = "o", markerfacecolor = "green", markersize = 9)

for i, txt in enumerate(First_Method_1["Iterations"]):

    plt.annotate(txt, (First_Method_1["Time"][i], First_Method_1["Loss"][i]), fontsize=20)

plt.xlabel("Time of Execution (microseconds)", fontsize=20); plt.ylabel("Loss (value of function G)", fontsize=20)

plt.legend(prop={'size':20})

plt.savefig(".\img\Fig1.eps", format = "eps")

plt.show()
```
