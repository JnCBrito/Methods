```python
def bootstrap_sample(x, Nboot=10):
    
    k = int(len(x)/2)
    bootstrap_samples_mean, bootstrap_samples_var = [], []
    for _ in range(Nboot):
        
        y = shuffle(x)
        sample = random.sample(y.tolist(), k)
        bootstrap_samples_mean.append(np.mean(sample))
        error = [pow(sample[i] - bootstrap_samples_mean[-1],2) for i in range(len(sample))]
        bootstrap_samples_var.append(sum(error)/(pow(k*4,2)*2))
        
    return [bootstrap_samples_mean,bootstrap_samples_var]
```

```python
def print_results_mean(k, mean, sample):
    
    for i in range(k):
    
        print('Expected value {0:1.0f}: {1:.2f}'.format(i+1,mean[i]))
        print('95% Confidence Interval {0:1.0f}: [{1:.2f} - {2:.2f}]'.format(
            i+1,np.percentile(sample[i],2.5),np.percentile(sample[i],97.5)
        ))
```

```python
def print_results_variance(k, var, sample):
    
    for i in range(k):
    
        print('Expected value {0:1.0f}: {1:.6f}'.format(i+1,var[i]))
        print('95% Confidence Interval {0:1.0f}: [{1:.6f} - {2:.6f}]'.format(
            i+1,np.percentile(sample[i],2.5),np.percentile(sample[i],97.5)
        ))
```

```python
bootstrap_means, bootstrap_var = [], []
for i in range(len(data)):
    
    aux = bootstrap_sample(data[i])
    bootstrap_means.append(aux[0])
    bootstrap_var.append(aux[1])
```
